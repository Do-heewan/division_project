```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.array([2, 3, 4, 5, 6, 7])
score = np.array([25, 49, 42, 57, 72, 68])
```

```python
plt.scatter(hours, score)
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
def model(x, w=3, b=35):
    return w*x +b

hours_0to10 = np.linspace(0, 10, 11)
h = model(hours_0to10)
print(h)
```

```python
plt.plot(hours_0to10, h, 'r')
plt.scatter(hours, score) # scatter plot
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
def cost(y, y_hat):
    return np.mean((y-y_hat)**2)

h = model(hours)
loss = cost(score, h)
print(loss)
```

```python
w = 3
b = 35

w_init = w
b_init = b
 
print(w)
print(b)
```

```python
#hyperparameter
lr = 0.001
epochs = 10000
```

```python
def model(x, w=w, b=b):
    return w*x +b
```

```python
n = hours.shape[0]

loss = []

for i in range(epochs):
    h = model(hours, w=w, b=b)
    
    loss.append(cost(score, h))
    dW = (2/n) * sum(hours * (h-score))
    db = (2/n) * sum(h-score)

    w = w - lr * dW
    b = b - lr * db
    
    print(loss[-1])
```

```python
plt.plot(loss)
plt.ylim(0,300)
```

```python
print(f'Before: w: {w_init}, b: {b_init})')
print(f'After: w: {w}, b: {b})')
```

```python
h      = model(hours_0to10, w=w,b=b)
h_init = model(hours_0to10, w=w_init, b=b_init)

plt.scatter(hours, score)
plt.plot(hours_0to10,h_init, label='Initial', color='r')
plt.plot(hours_0to10,h, label='Training', color='b')
plt.legend()
plt.show()
```
